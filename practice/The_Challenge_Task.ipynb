To assist them, we will be implementing a Python method that utilizes PCA (Principal Component Analysis) to reduce the dimensionality of the data. This reduction in dimensionality will help us visualize the data in a lower-dimensional space.



```python
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import seaborn as sns
import matplotlib.pyplot as plt
```

Loading the dataset

```python
data = pd.read_csv("Challenge_task_data.csv")
data.head()
```

Create two variables `X` and `y`. Seperate the feature columns and `Group` column.

Store the  feature columns in `X` and the `Group` column in `y`

```python
X = data.drop('Group', axis=1).values
y = data.Group
```

Write a method called `PCA_Visualization(features, labels)`:

* This method will apply a Principal Component Analysis to the `features` to reduce the data dimensionality to `2` features.
* Convert the transformed data into a dataframe
* Use seaborn scatter plot to visualize the data points.apply colors to different groups of data points `(hue = 'Group')`
* visualize the data using the method

```python
def PCA_Visualization(features, labels):
    pca = PCA(2)
    pca.fit(features)
    transformed_data = pca.transform(features)
    df = pd.DataFrame(transformed_data)
    df['Group'] = labels
    plt.figure(figsize=(10, 10))
    sns.scatterplot(df, x=0, y=1,hue='Group')

PCA_Visualization(X, y)
```
